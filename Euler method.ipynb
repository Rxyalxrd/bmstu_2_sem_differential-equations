# Метод Эйлера для решения обыкновенных дифференциальных уравнений (ODE)

Метод Эйлера представляет собой простой численный метод для решения обыкновенных дифференциальных уравнений (ODE). Данный метод позволяет найти приближенное решение ODE на основе заданного начального условия.

## Классы

### Euler

Класс `Euler` представляет собой реализацию метода Эйлера для решения обыкновенных дифференциальных уравнений.

#### Методы:

- `__init__(self, func, x0, y0, step, dx)`: Конструктор класса, инициализирующий объект метода Эйлера.
- `solution(self)`: Метод, который вычисляет приближенное решение ODE методом Эйлера.

#### Атрибуты:

- `func`: Функция правой части дифференциального уравнения.
- `x0`: Начальное значение аргумента.
- `y0`: Начальное значение функции.
- `step`: Количество шагов (итераций).
- `dx`: Шаг сетки.

### ODEEquals

Класс `ODEEquals` содержит статический метод `initial_func`, представляющий правую часть обыкновенного дифференциального уравнения (ODE).

#### Методы:

- `function(x, y)`: Статический метод, представляющий правую часть дифференциального уравнения.

## Использование

Для использования метода Эйлера для решения обыкновенных дифференциальных уравнений, создайте экземпляр класса `Euler`, передав соответствующую функцию правой части, начальные условия, количество шагов и шаг сетки. Затем вызовите метод `solution()` для получения решения.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
class Euler:

    def __init__(self, func, x_start, x_end, y0):
        self.func = func
        self.x_start = x_start
        self.x_end = x_end
        self.y0 = y0
    
    def solution(self, num_points):
        points = [(self.x_start, self.y0)]
        x, y = self.x_start, self.y0
        dx = (self.x_end - self.x_start) / num_points
        for _ in range(num_points):
            if x >= self.x_end:
                break
            y += dx * self.func(x, y)
            x += dx
            points.append((x, y))
        return np.array(points)
    
class ODEEquals:
    
    @staticmethod
    def function(x, y): 
        return (y - x + 1)/(y - x + 2)


euler_solver = Euler(func=ODEEquals.function, x_start=0, x_end=4, y0=1)

solution = euler_solver.solution(num_points=50)
```

```python
plt.plot(solution[:, 0], solution[:, 1], label="Euler method.")
plt.title(f"Solution of ODEs by Euler's method.")
plt.xlabel("x")
plt.ylabel("y")
plt.locator_params(axis="x", nbins=15) 
plt.locator_params(axis="y", nbins=30)  
plt.legend()
plt.show()
```
