```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class FourierSeries:
    def __init__(self, start: float = -1, end: float = 1, step: float = 1e-2, iterations: int = 30):
        self.start = start
        self.end = end
        self.step = step
        self.iterations = iterations
        self.x_values = np.arange(self.start, self.end, self.step)
        self.y_values = self.compute_series(self.x_values)

    def compute_series(self, x: np.ndarray) -> np.ndarray:
        y = np.zeros_like(x)
        a0 = (1/2) * np.trapz(x**3, x)
        y += a0 / 2
        for n in range(1, self.iterations + 1):
            a_n = np.trapz(x**3 * np.cos(n * np.pi * x), x)
            b_n = np.trapz(x**3 * np.sin(n * np.pi * x), x)
            y += a_n * np.cos(n * np.pi * x) + b_n * np.sin(n * np.pi * x)
        return y

    def plot(self):
        plt.plot(self.x_values, self.y_values, label='$S_n(x^3)$')
        plt.plot(self.x_values, self.x_values**3, label='$x^3$')
        plt.legend()
        plt.show()
```

```python
fourier_series = FourierSeries()
fourier_series.plot()
```
